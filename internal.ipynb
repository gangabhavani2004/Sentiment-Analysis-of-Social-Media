```python
import pandas as pd
import numpy as np

# Create a DataFrame with random data
data = np.random.randint(1, 100, size=(10, 10))
data
```

```python
df = pd.DataFrame(data,columns=['one','two','three','four','five','six','seven','eight','nine','ten'])
df
```

```python
# Arrange columns in ascending order
df_asc = df.apply(lambda x: x.sort_values().values, axis=0)
df_asc.columns = df.columns
print("DataFrame with Columns in Ascending Order:")
print(df_asc)
print("\n")

# Arrange columns in descending order
df_desc = df.apply(lambda x: x.sort_values(ascending=False).values, axis=0)
df_desc.columns = df.columns
print("DataFrame with Columns in Descending Order:")
print(df_desc)
print("\n")
```

```python


row_stats = df.apply(lambda row: pd.Series({'Mean': row.mean(), 'Median': row.median(), 'Mode': row.mode()[0]}), axis=1)
row_stats
```
